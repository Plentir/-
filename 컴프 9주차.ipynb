```python
#월요일 반 시험 답안지

pln = 0
p = []

for n in range(110, 1001):
    rev_n = str(n)[::-1]
    tot = n + int(rev_n)
    
    if tot == int(str(tot)[::-1]):
        pln = pln + n
        p.append(n)
        
print(p)
print(pln)
```

# 파일 생성하기
코드에서 사용하는 파일들의 디렉토리는 해당 코드의 디렉토리와 동일해야 한다.
한편, 코드를 이용해 생성한 파일의 디렉토리 또한 해당 코드의 디렉토리와 동일하다.

```python
f = open('new.txt', 'w')

# .write()는 구문 안에 문자열만 사용할 수 있다.
f.write('Hello, world! \n') # \n(역슬래시 = \로 보일 수 있음.)은 문자열 내에 넣어서 줄을 바꾸는 역할을 한다. 문자열 내에 있지만 문자로 출력되지 않는다.
f.write('Hello, python! \n')
f.write('--------------------------------------------------------------- \n')
f.write('한 줄에 \n이어서 써도 \n줄 바꿈이 작동한다.')

f.close() # .close()를 사용하지 않은 상태에서 파일을 읽거나 수정할 경우, 오류를 일으키기 때문에 한번 연 파일은 끝에 닫아주어야 한다.
```

```python
f = open('turtle.txt', 'w')

for i in range(1, 100 + 1):
    f.write('%3d번 거북이 출첵 \n'%i)

f.close()
```

```python
f = open('turtles.txt', 'a') # 'w'는 새로 쓰기의 의미이다. 'a'는 내용 추가하기이다.

f.write('100마리 거북이 출석')

f.close()
```

```python
f = open('one_mol.sdf', 'r')

mol = f.readlines() #파일 전체를 리스트화
info = mol[3].split() #리스트에서 세 번째 문자를 가져온 뒤 앞뒤 제거. 이 경우 파일에서의 각 행이 리스트의 요소로 들어가므로 mol[3]은 넷째 줄을 의미.
n_atm = info[0] #넷째 줄의 첫 번째 문자.

print(n_atm)
```

```python
f = open('one_mol.sdf', 'r')
atom_box = []

mole = f.readlines() #파일 전체를 리스트화
n_atom = mole[3].split()[0] #리스트에서 세 번째 문자를 가져온 뒤 앞뒤 제거. 이 경우 파일에서의 각 행이 리스트의 요소로 들어가므로 mol[3]은 넷째 줄을 의미.
atm_inf = mole[4 : 4 + int(n_atom)] #넷째 줄의 첫 번째 문자.
print(atm_inf)

for i in atm_inf:
    line = i.split()
    print(line[3])
    atom_box.append(atm_inf)

print(atom_box)
```

```python
f = open('one_mol.sdf', 'r')
atms = []
atm_w = {'H' : 1, 'C' : 12, 'N' : 14, 'O' : 16, 'F' : 19, 'Cl' : 35, 'Br' : 80}

mol = f.readlines()
n_atm = mol[3].split()[0] # 원자 번호.
inf_atm = mol[4 : 4 + int(n_atm)]

for i in inf_atm:
    line = i.split()
    atm = line[3]
    atms.append(atm)

w_mol = 0
for p in atms:
    w_atm = atm_w[p]
    w_mol = w_mol + w_atm
    
print(w_mol)
```
