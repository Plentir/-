```python
#110이상 1000이하의 자연수.
#숫자 120을 뒤집으면 021이 되는데 120과 21을 더하면 141이 되고, 141은 회문이다.
#같은 과정으로 회문이 되는 수 중, 주어진 범위 내에 있는 수의 총합을 구하라.

pal = []

for i in range(110, 1000 + 1):
    istr = str(i)
    irev = istr[::-1]
    irv = int(irev)
    a = i + irv
    al = list(str(a))
    
    if al == al[::-1]:
        pal.append(i)
        
print(pal)
#print(res)
```
